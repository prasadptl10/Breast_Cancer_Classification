```python
import pandas as pd
import numpy as np
import matplotlib 
import sklearn.datasets
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#Loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
#Loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

#Adding target column to the existing feature columns 

```python
data_frame['label']=breast_cancer_dataset.target
data_frame.tail()
```

```python
data_frame.shape
```

```python
data_frame.info()
```

```python
data_frame.isnull().sum()
```

```python
#Statistical measures about the data

data_frame.describe()
```

```python
#Checking the distribution of target variable

data_frame['label'].value_counts()
```

# What does label columns defines 
1 --> Benign
0 --> Malignant

```python
# Grouping the label and giving us the mean value

data_frame.groupby('label').mean()
```

Separating the features and target      

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and testing data

```python
X_train,X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardise the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

Building the Neural Network

```python
#Importing tensorflow and keras

import tensorflow as tf 
tf.random.set_seed(3)
from tensorflow import keras
```

```python
 # Setting up the layers of neural network

 model = keras.Sequential([
                            keras.layers.Flatten(input_shape=(30,)),                          # we have to flatten the 30 columns in 1D array as it acts as input neuron
                            keras.layers.Dense(20, activation = 'relu'),                      # we have 20 hidden neurons
                            keras.layers.Dense(2, activation = 'sigmoid')                     # we have 2 output neurons for cloass 0 & 1 
                            

 ])
```

```python
# Compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',    #Sparase when target variablesa are intergers or normal labelling 
              metrics = ['accuracy'])
```

```python
#  Training the neural network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)            #epochs = how many time the model go through the data
```

Visualizing the accuracy and the loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
